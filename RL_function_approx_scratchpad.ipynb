```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# builds (poly) features based on input data 
def poly_features(data,deg):
    F = []
    for n in range(deg+1):
        for m in range(deg+1):
            if n + m <= deg:
                temp = (data[:,0]**n)*(data[:,1]**m)
                temp.shape = (len(temp),1)
                F.append(temp)
    F = np.asarray(F)
    F.shape = (np.shape(F)[0],np.shape(F)[1])
    return F
```

```python
s = np.linspace(0,1)
xx,yy = np.meshgrid(s,s)
xx.shape = (np.prod(np.shape(xx)),1)
yy.shape = (np.prod(np.shape(yy)),1)
data = np.concatenate((xx,yy),axis = 1)

poly_data = poly_features(data = data,deg = 7)

w = np.random.randn(np.shape(poly_data)[0],1)
zz = np.dot(w.T,poly_data)

xx.shape = (len(s),len(s))
yy.shape = (len(s),len(s))
zz.shape = (len(s),len(s))

fig = plt.figure(num=None, figsize=(5,5), dpi=80, facecolor='w', edgecolor='k')
ax1 = plt.subplot(111,projection = '3d')
fig.subplots_adjust(left=0,right=1,bottom=0,top=1)   # remove whitespace around 3d figure
ax1.plot_surface(xx,yy,zz)
plt.show()
```
