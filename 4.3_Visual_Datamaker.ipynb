### Random linear classification datasets
The next cell imports the file and creates an instance of the datamaker. 

```python
%matplotlib inline
# import datamaker file
from random_classification_datamakers import Random_Classification_Datamaker

# create an instance of the datamaker
demo = Random_Classification_Datamaker()
```

The following cells demonstrate some random linear classification datasets. 

For example, create a random two-class dataset.

```python
# generate data, labels, and true linear separators: num_pts = number of points in dataset, num_seps = number of separators
demo.make_linear_classification_dataset(num_pts = 200,num_seps = 1)
```

```python
# generate data, labels, and true linear separators: num_pts = number of points in dataset, num_seps = number of separators
demo.make_linear_classification_dataset(num_pts = 200,num_seps = 2)
```

Or three linear boundaries...just set num_seps = 3.

```python
# generate data, labels, and true linear separators: num_pts = number of points in dataset, num_seps = number of separators
demo.make_linear_classification_dataset(num_pts = 200,num_seps = 3)
```

#  2. Make your own toy classification or clustering dataset by point and click action

The next few cells show you how to use another simple tool for creating multiclass toy datasets by visual point-and-click action.

Create an instance of the datamaker tool below, and an empty plot will appear.  Put points of one class wherever you like by clicking on the plotting area.  Once you are finished placing points from one class, click just outside of the square plotting area in the Figure to change classes, then place points from the next class - these should show in a different color than your previous class (maximum of 5 classes allowed).  And so on.  Once you have finished placing points call the 'save_data' function to save your data with labels.

Note: this can be used to make datasets for both classification and clustering experiments.

```python
# turn on interactive matplotlib
%matplotlib nbagg

# import datamaker library
from point_n_click_classification_datamaker import Classification_Datamaker

# create an instance of the datamaker
demo = Classification_Datamaker()
```

```python
# activate this cell to save the data you made abobve
demo.save_data(csvname = 'toy_clustering_data.csv')
```
